### Part B

From the global environment to a virtual environment to a container to a virtual computer, these systems get increasingly isolated from the device that you're working from.

The global environment is the default usage of storage and computation within your device, which is utilized by all software on your device unless otherwise specified.

A virtual environment is comprised of a folder within your device with a unique configuration of Python and its packages (as well as their respective versions), which can differ from your base Python in the global environment (slightly isolating it from the global environment).

A container is a virtual environment in which you can also control the operating system and install software outside of Python, so it is not limited to the operating system of your device (making it more isolated that the regular virtual environment).

A virtual machine is a separate virtual computer that you can connect to remotely and configure some of the hardware utilized. You can customize things like the GPU, CPU, and amount of memory utilized. A virtual machine is not confined to the hardware of your device as it is completely separate, so it is completely isolated from your device.

### Part B

```python
def collatz(n):
    if not isinstance(n, int):
        raise TypeError("Inputted number is not an integer!")
    if n < 0:
        raise ValueError("Inputted number is negative!")

    if n % 2 == 0:
        return n // 2
    else:
        return 3 * n + 1

print(collatz(4))  # Output: 2
```
