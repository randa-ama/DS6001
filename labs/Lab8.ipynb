## **Problem 2**

### **Part A**

```python
url = 'https://www.cdc.gov/workplace-health-promotion/media/files/2024/06/whpps_120717.csv'
df = pd.read_csv(url, delimiter='~')
df.head()
```

```python
df.info(verbose=True)
```

The data has 2,843 rows and 301 columns. There are 294 float columns and 7 object columns.

### **Part B**

```python
df.columns[df.columns.str.contains('WD')]
```

```python
whpps = df[['Industry','Size','OC3','HI1','HI2','HI3','CP1',
            'WL6','WD1_1', 'WD1_2', 'WD2', 'WD3', 'WD4', 'WD5',
            'WD6', 'WD7']]
whpps.head()
```

### **Part C**

```python
whpps = whpps.rename({'OC3': 'OrgStructure',
                      'HI1': 'InsFTEmp',
                      'HI2': 'InsFTPremiumChange',
                      'HI3': 'InsPTEmp',
                      'CP1': 'HealthEdOffered',
                      'WL6': 'WFHAllowed',
                      'WD1_1': 'PctEmpUnder30', 
                      'WD1_2': 'PctEmpOver60', 
                      'WD2': 'PctEmpFemale', 
                      'WD3': 'PctEmpHourly', 
                      'WD4': 'PctEmpNotDayShift', 
                      'WD5': 'PctEmpRemoteOffSite', 
                      'WD6': 'PctEmpUnion', 
                      'WD7': 'AvgYrTurnoverPct'}, axis = 1)

whpps.head()
```

### **Part D**

#### Step I

```python
replace_map = {
    1:'Food, Power, Utility Acquisition and Construction',
    2:'Trade and Logistics',
    3:'Leisure and Other',
    4:'Corporate Services',
    5:'Healthcare and Education',
    6:'Public Admin',
    7:'Hospital'
}

whpps['Industry'] = whpps['Industry'].replace(replace_map)
whpps['Industry'].value_counts()
```

#### Step II

```python
size_replace = {1: 'Small',
                2: 'Small',
                3: 'Small',
                4: 'Medium',
                5: 'Medium',
                6: 'Large',
                7: 'Large',
                8: 'Large'}

whpps['Size'] = whpps['Size'].replace(size_replace)
whpps['Size'].value_counts(dropna=False)
```

#### Step III

```python
whpps.dtypes
```

```python
replace_map = {'OrgStructure':{1:'For Profit, Public',
                               2:'For Profit, Private',
                               3:'Non-Profit',
                               4:'State or Local Government',
                               5:'Federal Government',
                               6:'Other',
                               97: np.nan,
                               98: np.nan,
                               99: np.nan},
                'InsFTEmp':  {1:'Full Insurance Offered',
                              2:'Partial Insurance Offered',
                              3:'No Insurance Offered',
                              97: np.nan,
                              98: np.nan,
                              99: np.nan},
                'InsFTPremiumChange': {1:'Larger',
                                       2:'Smaller',
                                       3:'About the Same',
                                       96: np.nan,
                                       97: np.nan,
                                       98: np.nan,
                                       99: np.nan},
                'InsPTEmp': {1:'Yes',
                             2:'No',
                             97: np.nan,
                             98: np.nan,
                             99: np.nan},
                'HealthEdOffered': {1:'Yes',
                                    2:'No',
                                    97: np.nan,
                                    98: np.nan},
                'WFHAllowed': {1:'Yes',
                               2:'No',
                               97: np.nan,
                               98: np.nan,
                               99: np.nan}
                }

whpps = whpps.replace(replace_map)
```

#### Step IV

```python
whpps.dtypes
```

```python
whpps.select_dtypes(include='float64').columns
```

```python
num_cols = whpps.select_dtypes(include='float64').columns

whpps[num_cols] = whpps[num_cols].replace([997,998,999],np.nan)
```

### **Part E**

```python
whpps = whpps.sort_values(by=['Industry','Size','PctEmpUnder30'],
                  ascending=[True,True,False])
# truncated for sake of submission
whpps.head(10)
```

### **Part F**

```python
whpps = whpps.query('Industry.isna() == False')
len(whpps)
```

### **Part G**

```python
whpps['GenderBalance'] = pd.cut(whpps.PctEmpFemale, 
                                bins=[-0.1,35,65,100],
                                labels = ('Mostly men','Balanced','Mostly women'))

whpps['GenderBalance'].value_counts()
```
