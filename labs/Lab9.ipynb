## **Problem 4**

### **Part A**

I expect this merge to be one-to-one, because both of the datasets have one row for each country and year combination, so ideally, all of the rows in one should match up to only one row in the other.

### **Part B**

```python
vdwb =  pd.merge(vdems, 
                  wbs,
                  on = ['country_code','year'],
                  how = 'outer',
                  validate = 'one_to_one',
                  indicator = 'matched')

vdwb.head()
```

### **Part C**

```python
vdwb['matched'].value_counts()
```

```python
# wbs only

wb_merge = vdwb.query("matched == 'right_only'")
wb_merge.groupby(['country_code','country_name_wb']).agg({'year':['min','max']})
```

```python
# vdems only

vdem_merge = vdwb.query("matched == 'left_only'")
vdem_merge.groupby(['country_code','country_name_vdem']).agg({'year':['min','max']})
```

## **Problem 5**

### **Part A**

```python
vdwb_final.groupby('country_name_vdem').agg({'democracy':'mean'}).sort_values(by='democracy',ascending=False)
```

The 5 countries with the highest average levels of democracy are Denmark, Sweden, Germany, Belgium, and Luxembourg, respectively. The 5 countries with the lowest average levels of democracy are Saudi Arabia, Qatar, United Arab Emirates, Oman, and Eritrea, respectively.

### **Part B**

```python
vdwb_final['educational_equality'] = pd.cut(vdwb_final.educational_equality, 
                                           bins=[-4,-2,-0.5,0.5,1.5,4],
                                           labels = ('extremely unequal','very unequal',
                                                     'somewhat unequal','relatively equal',
                                                     'equal'))
    
vdwb_final.groupby('educational_equality').agg({'gini_index':'mean',
                                                'poverty_headcount_ratio_at_national_poverty_lines': 'mean'})
```

Based on the table, it seems that as educational equality increases in the countries, the economic inequality decreases as well as the proportion of the population living in poverty. Basically, the more people are given a high quality basic education within the countries, the less of an economic disparity there is betwen the rich and the poor and the less people living in poverty.
